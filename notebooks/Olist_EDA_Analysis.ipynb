## Chapter 1 — Growth Quality Diagnosis
Strategic Question: Is Olist's growth real or just seasonal noise?  
Techniques: Revenue decomposition, cohort analysis, retention analysis  
Verdict: Growth is volume driven masking a critical retention crisis

```python
# ============================================
#  GROWTH QUALITY DIAGNOSIS
# ============================================

# Step 1 — Remove incomplete 2016 data (only 3 months — misleading)
analysis_data = monthly_revenue[monthly_revenue['order_year'].isin([2017, 2018])].copy()

# Step 2 — Revenue Decomposition
# Growth = Volume Effect (more orders) + Price Effect (higher spend per order)
analysis_data['prev_revenue'] = analysis_data['total_revenue'].shift(1)
analysis_data['prev_orders'] = analysis_data['total_orders'].shift(1)
analysis_data['prev_aov'] = analysis_data['avg_order_value'].shift(1)

# Volume Effect = Change in orders × Previous AOV
analysis_data['volume_effect'] = (
    (analysis_data['total_orders'] - analysis_data['prev_orders']) * 
    analysis_data['prev_aov']
)

# Price Effect = Change in AOV × Current orders
analysis_data['price_effect'] = (
    (analysis_data['avg_order_value'] - analysis_data['prev_aov']) * 
    analysis_data['total_orders']
)

# Revenue Growth
analysis_data['revenue_growth'] = analysis_data['total_revenue'] - analysis_data['prev_revenue']
analysis_data['growth_pct'] = analysis_data['total_revenue'].pct_change() * 100

# Drop first row
analysis_data = analysis_data.dropna()

# What % of growth is volume vs price driven?
total_volume_effect = analysis_data['volume_effect'].sum()
total_price_effect = analysis_data['price_effect'].sum()
total_growth = total_volume_effect + total_price_effect

print("=" * 60)
print("REVENUE DECOMPOSITION — GROWTH QUALITY DIAGNOSIS")
print("=" * 60)
print(f"\nVolume Effect (More Orders):  R${total_volume_effect:>12,.0f}  ({total_volume_effect/total_growth*100:.1f}%)")
print(f"Price Effect (Higher Spend):  R${total_price_effect:>12,.0f}  ({total_price_effect/total_growth*100:.1f}%)")
print(f"Total Growth:                 R${total_growth:>12,.0f}  (100%)")
print("\nDIAGNOSIS:")
if total_volume_effect > total_price_effect:
    print("   Growth is VOLUME DRIVEN — Olist needs MORE customers to grow")
    print("    Risk: Vulnerable to market saturation & high acquisition costs")
else:
    print("   Growth is VALUE DRIVEN — existing customers spending more")
    print("   Healthier: Less dependent on constant new customer acquisition")
print("=" * 60)
```

```python
# ============================================
# New vs Returning Customer Revenue Split
# ============================================

# Find customers with more than 1 order
customer_order_counts = master[master['order_status']=='delivered'].groupby(
    'customer_unique_id')['order_id'].nunique().reset_index()
customer_order_counts.columns = ['customer_unique_id', 'total_orders']

new_customers = customer_order_counts[customer_order_counts['total_orders'] == 1]
returning_customers = customer_order_counts[customer_order_counts['total_orders'] > 1]

total_customers = len(customer_order_counts)
retention_rate = len(returning_customers) / total_customers * 100

# Revenue from new vs returning
delivered = master[master['order_status'] == 'delivered'].copy()
delivered = delivered.merge(customer_order_counts, on='customer_unique_id')

new_revenue = delivered[delivered['total_orders'] == 1]['payment_value'].sum()
returning_revenue = delivered[delivered['total_orders'] > 1]['payment_value'].sum()
total_rev = new_revenue + returning_revenue

print("=" * 60)
print(" CUSTOMER RETENTION INTELLIGENCE")
print("=" * 60)
print(f"\n New Customers (1 order):      {len(new_customers):>8,}  ({len(new_customers)/total_customers*100:.1f}%)")
print(f" Returning Customers (2+):     {len(returning_customers):>8,}  ({retention_rate:.1f}%)")
print(f"\n Revenue from New Customers:   R${new_revenue:>12,.0f}  ({new_revenue/total_rev*100:.1f}%)")
print(f" Revenue from Returning:       R${returning_revenue:>12,.0f}  ({returning_revenue/total_rev*100:.1f}%)")
print(f"\n  RETENTION RATE: {retention_rate:.1f}%")
print("\n BUSINESS DIAGNOSIS:")
if retention_rate < 10:
    print("    CRITICAL: Retention is dangerously low!")
    print("   Olist is running a LEAKY BUCKET business model")
    print("   Constantly acquiring customers but not keeping them")
    lost_revenue = new_revenue * 0.05
    print(f"    Conservative Revenue Recovery Opportunity: R${lost_revenue:,.0f}")
    print("   if just 5% more customers make a second purchase")
elif retention_rate < 20:
    print("     WARNING: Retention needs improvement")
else:
    print("    Healthy retention rate")
print("=" * 60)
```

```python
# ============================================
# COHORT ANALYSIS 
# ============================================

# Get first purchase month for each customer
cohort_data = delivered[['customer_unique_id', 'order_purchase_timestamp']].copy()
cohort_data['order_month'] = cohort_data['order_purchase_timestamp'].dt.to_period('M')

# First purchase month = cohort
cohort_data['cohort'] = cohort_data.groupby(
    'customer_unique_id')['order_purchase_timestamp'].transform('min').dt.to_period('M')

# Months since first purchase
cohort_data['months_since_first'] = (
    cohort_data['order_month'] - cohort_data['cohort']
).apply(lambda x: x.n)

# Cohort size
cohort_size = cohort_data.groupby('cohort')['customer_unique_id'].nunique()

# Cohort retention matrix
cohort_matrix = cohort_data.groupby(
    ['cohort', 'months_since_first'])['customer_unique_id'].nunique().unstack()

# Retention rate
retention_matrix = cohort_matrix.divide(cohort_size, axis=0) * 100

# Plot cohort heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(
    retention_matrix.iloc[:18, :12],
    annot=True, fmt='.0f',
    cmap='RdYlGn',
    linewidths=0.5,
    annot_kws={'size': 8},
    vmin=0, vmax=10
)
plt.title(' Customer Cohort Retention Matrix (%)\nHow many customers return each month after first purchase?',
          fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Months Since First Purchase', fontsize=12)
plt.ylabel('Cohort (First Purchase Month)', fontsize=12)
plt.tight_layout()
plt.savefig(r'C:\Users\ANUSREE MANJULA\OneDrive\Desktop\Olist_EDA_Project\visuals\cohort_retention.png',
            dpi=150, bbox_inches='tight')
plt.show()
print("Cohort Analysis Complete!")
```

## Chapter 2 — Delivery Intelligence
Strategic Question: At what exact point do late deliveries destroy customer trust?

## Chapter 4 — Customer Value Intelligence
Strategic Question: Are high spending customers actually Olist's most valuable ones?  
Techniques: RFM segmentation, customer lifetime value, retention analysis  
Verdict: Only 6.9% champions while 23.8% of customers are already at risk

```python
# calculating rfm metrics for each customer
# recency = how recently they purchased
# frequency = how many times they purchased  
# monetary = how much they spent in total

import datetime

# using the most recent date in dataset as reference point
reference_date = delivered_only['order_purchase_timestamp'].max()

rfm = delivered_only.groupby('customer_unique_id').agg(
    recency=('order_purchase_timestamp', lambda x: (reference_date - x.max()).days),
    frequency=('order_id', 'nunique'),
    monetary=('payment_value', 'sum')
).reset_index()

print("RFM Base Metrics")
print("-" * 40)
print(f"Total customers analyzed: {len(rfm):,}")
print(f"Average recency: {rfm['recency'].mean():.0f} days since last purchase")
print(f"Average frequency: {rfm['frequency'].mean():.2f} orders per customer")
print(f"Average monetary value: R${rfm['monetary'].mean():,.2f}")
print("-" * 40)
print(rfm.describe().round(2))
```

```python
# scoring each customer on recency, frequency and monetary value
# using quintiles so each score is relative to the entire customer base

rfm['r_score'] = pd.qcut(rfm['recency'], q=5, labels=[5,4,3,2,1])
rfm['f_score'] = pd.qcut(rfm['frequency'].rank(method='first'), q=5, labels=[1,2,3,4,5])
rfm['m_score'] = pd.qcut(rfm['monetary'], q=5, labels=[1,2,3,4,5])

rfm['r_score'] = rfm['r_score'].astype(int)
rfm['f_score'] = rfm['f_score'].astype(int)
rfm['m_score'] = rfm['m_score'].astype(int)

# combining scores into single rfm score
rfm['rfm_score'] = rfm['r_score'].astype(str) + rfm['f_score'].astype(str) + rfm['m_score'].astype(str)
rfm['total_score'] = rfm['r_score'] + rfm['f_score'] + rfm['m_score']

# segmenting customers based on rfm score combinations
def assign_segment(row):
    r = row['r_score']
    f = row['f_score']
    m = row['m_score']
    total = row['total_score']
    
    if r >= 4 and f >= 4 and m >= 4:
        return 'champion'
    elif r >= 3 and f >= 3 and m >= 3:
        return 'loyal customer'
    elif r >= 4 and f <= 2:
        return 'promising new customer'
    elif r <= 2 and f >= 3:
        return 'at risk customer'
    elif r <= 2 and f <= 2 and m >= 3:
        return 'high value lost customer'
    elif r == 1 and f == 1:
        return 'lost customer'
    else:
        return 'needs attention'

rfm['segment'] = rfm.apply(assign_segment, axis=1)

# segment summary
segment_summary = rfm.groupby('segment').agg(
    customer_count=('customer_unique_id', 'count'),
    avg_recency=('recency', 'mean'),
    avg_frequency=('frequency', 'mean'),
    avg_monetary=('monetary', 'mean'),
    total_revenue=('monetary', 'sum')
).reset_index()

segment_summary['revenue_share'] = (
    segment_summary['total_revenue'] / 
    segment_summary['total_revenue'].sum() * 100
)

segment_summary = segment_summary.sort_values('total_revenue', ascending=False)

print("RFM Customer Segmentation Results")
print("-" * 40)
print(segment_summary.to_string(index=False))
```

Techniques: BCG matrix classification, growth rate analysis, category lifecycle detection

```python
# adding time columns to delivered_only since they were calculated on master
delivered_only['order_year'] = delivered_only['order_purchase_timestamp'].dt.year
delivered_only['order_month'] = delivered_only['order_purchase_timestamp'].dt.month

# calculating category level metrics
category_metrics = delivered_only.groupby('product_category_name_english').agg(
    total_revenue=('payment_value', 'sum'),
    total_orders=('order_id', 'nunique'),
    avg_review_score=('review_score', 'mean'),
    avg_price=('price', 'mean')
).reset_index()

category_metrics = category_metrics.dropna(subset=['product_category_name_english'])

# growth rate - last 6 months vs previous 6 months
recent_cutoff = delivered_only['order_purchase_timestamp'].max() - pd.DateOffset(months=6)
mid_cutoff = delivered_only['order_purchase_timestamp'].max() - pd.DateOffset(months=12)

recent_orders = delivered_only[
    delivered_only['order_purchase_timestamp'] >= recent_cutoff
].groupby('product_category_name_english')['order_id'].nunique()

previous_orders = delivered_only[
    (delivered_only['order_purchase_timestamp'] >= mid_cutoff) & 
    (delivered_only['order_purchase_timestamp'] < recent_cutoff)
].groupby('product_category_name_english')['order_id'].nunique()

growth_rate = ((recent_orders - previous_orders) / previous_orders * 100).reset_index()
growth_rate.columns = ['product_category_name_english', 'growth_rate']

category_metrics = category_metrics.merge(growth_rate, on='product_category_name_english', how='left')

category_metrics['market_share'] = (
    category_metrics['total_revenue'] / 
    category_metrics['total_revenue'].max() * 100
)

category_metrics = category_metrics[category_metrics['total_orders'] >= 50]

print("Category Portfolio Overview")
print("-" * 40)
print(f"Total categories analyzed: {len(category_metrics)}")
print(f"Total revenue: R${category_metrics['total_revenue'].sum():,.0f}")
print("-" * 40)
print("Top 10 categories by revenue")
print(category_metrics.nlargest(10, 'total_revenue')[
    ['product_category_name_english', 'total_revenue', 
     'total_orders', 'growth_rate', 'avg_review_score']
].to_string(index=False))
```

```python
# classifying categories into bcg quadrants
# x axis = market share, y axis = growth rate

growth_median = category_metrics['growth_rate'].median()
share_median = category_metrics['market_share'].median()

def bcg_quadrant(row):
    high_growth = row['growth_rate'] > growth_median
    high_share = row['market_share'] > share_median
    
    if high_growth and high_share:
        return 'star'
    elif not high_growth and high_share:
        return 'cash cow'
    elif high_growth and not high_share:
        return 'question mark'
    else:
        return 'dog'

category_metrics['bcg_quadrant'] = category_metrics.apply(bcg_quadrant, axis=1)

bcg_summary = category_metrics.groupby('bcg_quadrant').agg(
    category_count=('product_category_name_english', 'count'),
    total_revenue=('total_revenue', 'sum'),
    avg_growth=('growth_rate', 'mean'),
    avg_score=('avg_review_score', 'mean')
).reset_index()

print("BCG Matrix Summary")
print("-" * 40)
print(bcg_summary.to_string(index=False))
print("-" * 40)
print("Star Categories - invest and scale")
print(category_metrics[category_metrics['bcg_quadrant']=='star'][
    ['product_category_name_english','total_revenue','growth_rate']
].sort_values('total_revenue', ascending=False).head(5).to_string(index=False))
print()
print("Dog Categories - consider exiting")
print(category_metrics[category_metrics['bcg_quadrant']=='dog'][
    ['product_category_name_english','total_revenue','growth_rate']
].sort_values('total_revenue', ascending=True).head(5).to_string(index=False))
```

```python
import pandas as pd
import os

raw_path = r'C:\Users\ANUSREE MANJULA\OneDrive\Desktop\Olist_EDA_Project\data\raw'
cleaned_path = r'C:\Users\ANUSREE MANJULA\OneDrive\Desktop\Olist_EDA_Project\data\cleaned'

orders = pd.read_csv(os.path.join(raw_path, 'olist_orders_dataset.csv'))
order_items = pd.read_csv(os.path.join(raw_path, 'olist_order_items_dataset.csv'))
customers = pd.read_csv(os.path.join(raw_path, 'olist_customers_dataset.csv'))
payments = pd.read_csv(os.path.join(raw_path, 'olist_order_payments_dataset.csv'))
reviews = pd.read_csv(os.path.join(raw_path, 'olist_order_reviews_dataset.csv'))
products = pd.read_csv(os.path.join(raw_path, 'olist_products_dataset.csv'))
sellers = pd.read_csv(os.path.join(raw_path, 'olist_sellers_dataset.csv'))
category = pd.read_csv(os.path.join(raw_path, 'product_category_name_translation.csv'))

date_columns = ['order_purchase_timestamp', 'order_approved_at',
                'order_delivered_carrier_date', 'order_delivered_customer_date',
                'order_estimated_delivery_date']
for col in date_columns:
    orders[col] = pd.to_datetime(orders[col])

master = orders.merge(customers, on='customer_id', how='left')
master = master.merge(order_items, on='order_id', how='left')
master = master.merge(payments, on='order_id', how='left')
master = master.merge(products, on='product_id', how='left')
master = master.merge(sellers, on='seller_id', how='left')
master = master.merge(reviews[['order_id','review_score']], on='order_id', how='left')
master = master.merge(category, on='product_category_name', how='left')

delivered_only = master[master['order_status'] == 'delivered'].copy()
delivered_only['delivery_days'] = (delivered_only['order_delivered_customer_date'] - delivered_only['order_purchase_timestamp']).dt.days
delivered_only['delivery_delay_days'] = (delivered_only['order_delivered_customer_date'] - delivered_only['order_estimated_delivery_date']).dt.days
delivered_only['order_year'] = delivered_only['order_purchase_timestamp'].dt.year
delivered_only = delivered_only.dropna(subset=['delivery_days', 'delivery_delay_days', 'review_score'])

seller_metrics = pd.read_csv(os.path.join(cleaned_path, 'seller_metrics.csv'))
category_metrics = pd.read_csv(os.path.join(cleaned_path, 'category_metrics.csv'))

print("master and delivered_only rebuilt successfully")
```

```python
seller_yearly = delivered_only.groupby(['seller_id', 'order_year']).agg(
    total_revenue=('payment_value', 'sum'),
    total_orders=('order_id', 'nunique'),
    avg_review_score=('review_score', 'mean'),
    late_deliveries=('delivery_delay_days', lambda x: (x > 0).sum()),
    total_deliveries=('delivery_delay_days', 'count')
).reset_index()

seller_yearly['late_delivery_rate'] = (seller_yearly['late_deliveries'] / seller_yearly['total_deliveries'] * 100)
seller_yearly = seller_yearly.merge(seller_metrics[['seller_id', 'risk_category']], on='seller_id', how='left')
seller_yearly.to_csv(os.path.join(cleaned_path, 'seller_metrics_yearly.csv'), index=False)

category_yearly = delivered_only.groupby(['product_category_name_english', 'order_year']).agg(
    total_revenue=('payment_value', 'sum'),
    total_orders=('order_id', 'nunique'),
    avg_review_score=('review_score', 'mean')
).reset_index()
category_yearly = category_yearly.merge(category_metrics[['product_category_name_english', 'bcg_quadrant']], on='product_category_name_english', how='left')
category_yearly.to_csv(os.path.join(cleaned_path, 'category_metrics_yearly.csv'), index=False)

state_yearly = delivered_only.groupby(['customer_state', 'order_year']).agg(
    total_revenue=('payment_value', 'sum'),
    total_orders=('order_id', 'nunique'),
    avg_review_score=('review_score', 'mean'),
    late_deliveries=('delivery_delay_days', lambda x: (x > 0).sum()),
    total_deliveries=('delivery_delay_days', 'count')
).reset_index()
state_yearly['late_delivery_rate'] = (state_yearly['late_deliveries'] / state_yearly['total_deliveries'] * 100)
state_yearly.to_csv(os.path.join(cleaned_path, 'state_metrics_yearly.csv'), index=False)

print("yearly exports saved")
```
